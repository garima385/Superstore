### Transforming and modifying data (filtering, formatting, authenticating etc data )

```python
pattern = '|'.join(['ern US', ' US'])
companies['region']=companies['region'].str.replace(pattern,'')
print(companies)
```

```python
west_sales['quantity']=west_sales['quantity'].astype(str)
```

```python
west_sales['quantity']=west_sales['quantity'].str.replace(r"[a-zA-Z]",'')
west_sales['quantity']=west_sales['quantity'].astype(int)
```

```python
print(west_sales[west_sales.product_id.str.contains(r'[^-0-9a-zA-Z]')])
```

```python
east_sales['product_id']=east_sales['product_id'].str.replace(r"[^-0-9a-zA-Z]+","")
```

```python
west_sales['product_id']=west_sales['product_id'][~west_sales.product_id.str.contains('Unknown')]
west_sales.dropna(subset=['product_id'],inplace=True)
print(west_sales['product_id'])
```

```python
print(east_sales[east_sales.sales.str.contains(r'[$]')])
```

```python
east_sales['sales']=east_sales['sales'].str.replace(r"[$]",'')
east_sales['sales']=east_sales['sales'].astype(float)
```

```python
south_sales['Date']=pd.to_datetime(south_sales['Date'])
```

```python
south_sales['Date Shipped']=pd.to_datetime(south_sales['Date Shipped'])
```

```python
south_sales['Profit']=south_sales['Profit'].str.replace(r"[$]",'')
south_sales['Profit']=south_sales['Profit'].astype(float)
```

```python
south_sales.rename(index=str,columns={'Customer Number':'Customer ID','Date':'Order Date','Date Shipped':'Ship Date',
                                      'Order':'Order ID','Postal Code 1':'Postal Code','Shipping Method':'Ship Mode'}
                   ,inplace=True)
```

```python
east_sales['ord_date']=pd.to_datetime(east_sales['ord_date'],errors='coerce')
```

```python
east_sales.dropna(subset=['ord_date'],inplace=True)
```

```python
east_sales['ship_date']=pd.to_datetime(east_sales['ship_date'],errors='coerce')
```

```python
east_sales.dropna(subset=['ship_date'],inplace=True)
```

```python
east_sales.rename(index=str,columns={'order_number':'Order ID','customer_key':'Customer ID','ord_date':'Order Date',
                                     'ship_date':'Ship Date','post_code':'Postal Code','product_id':'Product ID',
                                     'customer_section':'Segment','ship_type':'Ship Mode','sales':'Sales','#_items':'Quantity',
                                    'reduction':'Discount','total':'Profit'}
                   ,inplace=True)
```

```python
west_sales['ship_date']=pd.to_datetime(west_sales['ship_date'],errors='coerce')
```

```python
west_sales['order_date']=pd.to_datetime(west_sales['order_date'],errors='coerce')
```

```python
west_sales.dropna(subset=['ship_date'],inplace=True)
```

```python
west_sales['postal_code']=west_sales['postal_code'].astype(int)
```

```python
east_sales['Quantity']=east_sales['Quantity'].str.replace(r"[a-zA-Z]",'')
east_sales['Quantity']=east_sales['Quantity'].astype(int)
```

```python
west_sales['sales']=west_sales['sales'].astype(float)
west_sales['discount']=west_sales['discount'].astype(float)
west_sales['profit']=west_sales['profit'].astype(float)
```

```python
west_sales.rename(index=str,columns={'order_id':'Order ID','customer_id':'Customer ID','order_date':'Order Date',
                                     'ship_date':'Ship Date','postal_code':'Postal Code','product_id':'Product ID',
                                     'segment':'Segment','ship_mode':'Ship Mode','sales':'Sales','quantity':'Quantity',
                                    'discount':'Discount','profit':'Profit'}
                   ,inplace=True)
```

### Loading transformed data into one file

```python
cs_sales=pd.merge(central_sales,south_sales,how='outer')
we_sales=pd.merge(west_sales,east_sales,how='outer')
```

```python
cs_sales.head(n=10)
```

```python
we_sales.head(n=10)
```

```python
all_sales=pd.merge(we_sales,cs_sales,how='outer')
```

```python
all_sales.head(n=10)
```

```python
products.rename(index=str,columns={'product_id':'Product ID','product_name':'Product Name','category':'Product Type'}
                   ,inplace=True)
```

```python
all_sales_prod=pd.merge(all_sales,products,how='left')
```

```python
all_sales_prod.head(n=10)
```

```python
print(all_sales_prod[all_sales_prod.isna().any(axis=1)])
```

```python
product_types.rename(index=str,columns={'product_type':'Product Type','product_category':'Product Category'}
                   ,inplace=True)
```

```python
all_sales_prod_cat=pd.merge(all_sales_prod,product_types,how='left')
```

```python
all_sales_prod_cat.head(n=10)
```

```python
postal_codes.rename(index=str,columns={'postal_code':'Postal Code','state':'State','region':'Region',
                                       'country':'Country'}
                   ,inplace=True)
```

```python
all_sales_prod_cat_post=pd.merge(all_sales_prod_cat,postal_codes,how='left')
```

```python
companies.rename(index=str,columns={'company_id':'Company ID','company':'Company','region':'Region',
                                       'contact':'Contact Person'}
                   ,inplace=True)
```

```python
superstore_sales=pd.merge(all_sales_prod_cat_post,companies,how='left')
```

```python
print(superstore_sales[superstore_sales.isna().any(axis=1)])
```

```python
superstore_sales.dropna(subset=['Product Name'],inplace=True)
```

```python
superstore_sales.head(n=10)
```

```python
superstore_sales['Ship Date']=superstore_sales['Ship Date'].dt.date
```

```python
superstore_sales['Order Date']=superstore_sales['Order Date'].dt.date
```
